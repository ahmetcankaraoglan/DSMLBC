```python
#Gerekli import işlemleri ve veri setinin okutulması işlemleri gerçekleştirilmiştir.
import pandas as pd
import numpy as np
df = pd.read_csv("Automobile_data.csv")
```

```python
#Son 5 gözlem birimine erişilmiştir. 
df.tail()
```

**SORU 2**

```python
#Veri setinde eksik gözlem olup olmadığı incelenmiştir.
df.isnull().sum()
```

```python
#Veri setindeki eksik gözlemler silindi.
df = df.dropna()
```

```python
#Silme işlemi sonrasında tekrar eksik gözlem sorgulaması yapıldı.
df.isnull().sum()
```

```python
#Eksik gözlem olmadığı kesinleşince veri setimizi güncelledik.
df.to_csv("Automobile_data.csv")
```

```python
#İlk 5 gözlem birimine erişilmiştir. 
df.head()
```

**SORU 3**

```python
#En pahalı otomobil şirketinin adını bulun
df.groupby(["company"]).agg({"price":"max"}).sort_values("price", ascending = False).head(1)
```

**SORU 4**

```python
#Tüm Toyota Otomobillerin detaylarını yazdırın.
df[df['company'].str.contains("toyota")]
```

**SORU5**

```python
#Şirket başına düşen toplam araba sayısını bulun.
df["company"].value_counts()
```

**SORU 6**

```python
#Her şirketin en pahalı arabasını bulun.
df.groupby("company").agg({"company": "max","price":"max"})
```

**SORU7**

```python
#Her şirketin ortalama kilometresini bulun. 
df.groupby(["company"]).agg({"average-mileage":"mean"})
```

**SORU 8**

```python
#Tüm arabaları price değişkenine göre sıralayın.
df.sort_values("price", ascending = False).head()
```

**SORU 9**

```python
#Verilen iki veriyi alt alta birleştirin.
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925, 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi'], 'Price': [29995, 23600, 61500, 58900]}
```

```python
df1 = pd.DataFrame(GermanCars)
df2 = pd.DataFrame(japaneseCars)
```

```python
df3 = pd.concat([df1, df2], axis = 0, keys = ["Germany", "Japan"])
df3
```

**SORU 10**

```python
#Verilen iki veriyi yanyana ekleyin ve ikinci veride yer alan horsepower değişkenini yeni bir sutün olarak ekleyin.
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

```python
df1 = pd.DataFrame(Car_Price)
df2 = pd.DataFrame(car_Horsepower)
```

```python
pd.merge(df1, df2)
```
